```python
from geopy.geocoders import Nominatim
```

```python
locator = Nominatim(user_agent='myGeocoder')
```

```python
address=['1109 Shady Ln in Austin (TX)',
 '615 E. Wonsley in Austin (TX)',
 '124 W Anderson in Austin (TX)',
 '3118 Windsor Rd in Austin (TX)',
 '8238 Research Blvd in Austin (TX)',
 '4800 Weletka Dr in Austin (TX)',
 '183 And Cameron in Austin (TX)',
 '6808 S Ih 35 Frontage Rd in Austin (TX)',
 '8413 Danville Dr in Austin (TX)']
```

```python
lst=[]
for i in address:
    location = locator.geocode(i)
    lst.append(location)
    print('Generating location: '+i)
    
print('-----------------------------------')
print("--------------Done-----------------")
lst
```

```python

from geopy.extra.rate_limiter import RateLimiter
```

```python
df_test=pd.DataFrame(data=df['found_location']).iloc[0:5,:]
df_test
```

```python
geocode = RateLimiter(locator.geocode, min_delay_seconds=1)
# 2- - create location column
df_test['location'] = df_test['found_location'].apply(geocode)
# 3 - create longitude, laatitude and altitude from location column (returns tuple)
df_test['point'] = df_test['location'].apply(lambda loc: tuple(loc.point) if loc else None)
# 4 - split point column into latitude, longitude and altitude columns
# df_test[['latitude', 'longitude', 'altitude']] = pd.DataFrame(df_test['point'].tolist(), index=df_test.index)
```
