```python
from geopy.geocoders import Nominatim
```

```python
geolocator = Nominatim(user_agent="geoapiExercises")
```

```python
address=df['found_location'].tolist()
address
```

```python
lst=[]
for i in address:
    location = geolocator.geocode(i)
    lst.append(location)
    print('Generating location: '+i)
    
print('-----------------------------------')
print("--------------Done-----------------")
lst
```
